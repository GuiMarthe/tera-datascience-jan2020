# Análise exploratória de dados

Em estatística, a análise exploratória de dados (AED) é uma abordagem à análise de conjuntos de dados de modo a resumir suas características principais, frequentemente com métodos visuais. 
Primariamente a AED tem como objetivo observar o que os dados podem nos dizer além da modelagem formal ou do processo de teste de hipóteses. 

A análise exploratória de dados emprega grande variedade de técnicas gráficas e quantitativas, visando maximizar a obtenção de informações ocultas na sua estrutura, descobrir variáveis importantes em suas tendências, detectar comportamentos anômalos do fenômeno, testar se são válidas as hipóteses assumidas, escolher modelos e determinar o número ótimo de variáveis.

## Objetivos principais:
1. Entender suas variáveis
2. Limpar seu dataset
3. Analisar relação entre as variáveis

### Criar agrupamentos pode ajudar a achar padrões

````python
grupos = paises.groupby('Region')[['Country','Population','Area (sq. mi.)', 'Literacy (%)']].agg({'Country':'count', 'Population':'sum','Area (sq. mi.)':'sum', 'Literacy (%)':'mean'})
grupos
````

## Representação gráfica

````python
plt.figure(figsize=(16, 6))
sns.distplot(paises['GDP ($ per capita)']);
````

````python
plt.figure(figsize=(16, 6))
sns.distplot(paises['Infant mortality (per 1000 births)']);
````

````python
ax = sns.boxplot(x = 'Infant mortality (per 1000 births)', data = paises, orient = 'h')
ax.figure.set_size_inches(12, 4)
ax.set_title('Infant mortality (per 1000 births)', fontsize=18)
ax
````

````python
ax = sns.boxplot(x = 'Infant mortality (per 1000 births)', y='Region', data = paises, orient = 'h')
ax.figure.set_size_inches(12, 8)
ax.set_title('Infant mortality (per 1000 births)', fontsize=18)
ax
````

````python
plt.figure(figsize=(16, 6))
chart = sns.barplot(data=paises.groupby('Region')[['GDP ($ per capita)']].mean().sort_values(by='GDP ($ per capita)').reset_index(), 
            x = 'Region', y='GDP ($ per capita)')
plt.setp(chart.get_xticklabels(), rotation=90)
plt.show()
````

````python
plt.figure(figsize=(16, 6))
chart = sns.barplot(data=paises.groupby('Region')[['Population']].sum().sort_values(by='Population').reset_index(), 
            x = 'Region', y='Population')
plt.setp(chart.get_xticklabels(), rotation=90)
plt.show()
````

````python
plt.figure(figsize=(16, 6))
chart = sns.barplot(data=paises.groupby('Region')[['Infant mortality (per 1000 births)']].sum().sort_values(by='Infant mortality (per 1000 births)').reset_index(), 
            x = 'Region', y='Infant mortality (per 1000 births)')
plt.setp(chart.get_xticklabels(), rotation=90)
plt.show()
````

# Quais hipóteses podemos levantar analisando essas relações?

````python
sns.pairplot(paises, vars= ['GDP ($ per capita)','Infant mortality (per 1000 births)', 'Birthrate','Agriculture'],hue='Region',diag_kind="hist")
````

````python
sns.pairplot(paises, vars= ['GDP ($ per capita)','Phones (per 1000)', 'Literacy (%)','Service'],hue='Region',diag_kind="hist")
````
